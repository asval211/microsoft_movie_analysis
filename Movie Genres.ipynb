```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
conn = sqlite3.connect('data\movies.db')
```

```python
#Printing the query dataframe
query = pd.read_sql('''SELECT * FROM imdb_title_basics''', conn)
query
```

```python
#Joining Movie Gross table and Title Basics table 
query1 = pd.read_sql('''SELECT g.title, g.studio, g.domestic_gross, g.foreign_gross, g.year, t.genres
                    FROM bom_movie_gross AS g
                    JOIN imdb_title_basics AS t
                    USING(idx)''', conn)
query1
```

```python
#Converting foreign_gross column from string to integer
query1['foreign_gross'] = pd.to_numeric(query1['foreign_gross'], errors = 'coerce')
query1
```

```python
#Adding total_gross column
query1['total_gross'] = query1['domestic_gross'] + query1['foreign_gross']
query1
```

```python
#Top 10 Domestic Gross Films
query1.sort_values('domestic_gross', ascending=False)[:10]
```

```python
#Top 10 Foreign Gross Films
query1.sort_values('foreign_gross', ascending=False)[:10]
```

```python
#Top 10 Total Gross Films
query1.sort_values('total_gross', ascending=False)[:10]
```
