```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
conn = sqlite3.connect('data\movies.db')
```

```python
#Printing the query dataframe and rmeoving all rows with 0 in foreign_gross column
query = pd.read_sql('''SELECT * FROM bom_movie_gross WHERE foreign_gross IS NOT NULL''', conn)
query
```

```python
query.info()
```

```python
#Converting foreign_gross column from string to integer
query['foreign_gross'] = pd.to_numeric(query['foreign_gross'], errors = 'coerce')
query
```

```python
#Adding total_gross column
query['total_gross'] = query['domestic_gross'] + query['foreign_gross']
query
```

```python
#Top 10 Domestic Gross Films
query.sort_values('domestic_gross', ascending=False)[:10]
```

```python
#Top 10 Foreign Gross Films
query.sort_values('foreign_gross', ascending=False)[:10]
```

```python
#Top 10 Total Gross Films
query.sort_values('total_gross', ascending=False)[:10]
```
